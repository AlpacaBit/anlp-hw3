```python
bug_types = ['misused == or =', 'missing colons', 'unclosed parentheses', 'illegal separation', 
             'illegal indentation', 'unclosed string', 'illegal comment', 'faulty indexing', 'undefined objects', 
             'undefined methods', 'illegal keywords', 'condition error', 'operation error', 'variable error', 
             'other error', 'double', 'triple', 'quadruple']

correct_counts = [0 for i in range(len(bug_types))]
total_counts = [0 for i in range(len(bug_types))]

for file in files:
    # print(file)
    total = 0
    correct = 0
    bug_type = file.split("_")[1][:-5]
    ind = bug_types.index(bug_type)
    
    with open("res/gpt-35-turbo/debug/" + file, "r") as f:
        results = json.load(f)
        for result in results:
            total += 1
            
            if result["test_result_dict"] == None:
                continue
            
            elif "status_msg" in result["test_result_dict"].keys():
                status = result["test_result_dict"]["status_msg"]
                if status == "Accepted":
                    correct += 1
    
    correct_counts[ind] += correct
    total_counts[ind] += total
```

```python
pass_rates = [ ]
for i in range(len(correct_counts)):
    pass_rates.append(round(correct_counts[i] / total_counts[i] * 100, 2))
```

```python
pd.DataFrame({"bug type": bug_types, "pass rate": pass_rates, "correct": correct_counts, "total": total_counts})
```
